#자율주행
* 자율주행 자동차
자율주행 자동차는 운전자가 차량을 조작하지 않아도 스스로 움직이는 자동차를 말합니다. 국제 자동차기술자협회에 따르면 자율주행 자동차는 주행능력별로 레벨0에서 레벨5까지 6단계로 나뉘는데요. 레벨0에서 레벨2까지는 시스템이 일부 주행을 도와주는 단계로, 운전자가 주행의 주체가 되는 과거 및 현재의 자동차를 말합니다. 레벨0은 주행 중 안전을 위해 시스템은 단순 경고를 하거나 일시적으로 개입하는  수준으로서, ‘전방 충돌방지 보조’나 ‘후측방 충돌경고’ 등의 기능만을 갖춘 자율주행 기능이 없는 ‘일반 차량’의 단계입니다.

레벨1은 운전자를 보조하는 단계로, 특정 주행모드에서 시스템이 조향 또는 감·가속 중 하나를 수행하는 차로 유지보조(LFA), 스마트 크루즈 컨트롤(SCC) 등이 해당되며, 운전 시에는 운전자의 선택적인 제어가 필요합니다. 레벨2는 부분 자동화 단계로, 자동 스티어링 기능과 자동 속도조절 기능이 모두 추가된 단계인데요. 특정 주행모드에서 시스템이 조향 및 감·가속을 모두 수행하는 고속도로 주행 보조(HDA) 기능이 이에 해당되며, 이 또한 운전자의 선택적인 제어가 필요합니다. 현재 일반적으로 상용화되고 있는 차량의 자율주행기능은 레벨1에서 레벨2 단계입니다.

레벨3부터 레벨5까지는 시스템이 직접 주행을 수행하는 단계인데요. 레벨3은 조건부 자동화 단계로, 시스템이 차량 제어와 주행환경을 동시에 인식하며, 일부 특정 도로구간에서 자율주행이 가능한 단계입니다. 하지만 돌발상황 시에는 운전 제어권이 운전자에게 전환됩니다. 레벨4는 고도 자동화 단계로 시스템이 전체 주행을 하며, 긴급상황 발생 시에도 차량 스스로 대처가 가능한데요. 일부 특정 도로구간에서만 자율주행이 가능한 제약이 있습니다. 레벨5는 완전 자동화 단계로, 모든 도로와 교통조건에서 자율주행이 가능한 궁극적 ‘완전자율주행’이라고 할 수 있습니다.

지금까지 자율주행기능의 6단계에 대해 살펴봤는데요. 현재 전세계 자동차 업계는 2020년부터 2021년까지 레벨3 자율주행 차량의 양산화를 목표로 개발에 힘쓰고 있습니다. 아우디 A8, 캐딜락 슈퍼크루주, 테슬라 오토파일럿 등 레벨3 자율주행기능을 탑재한 모델을 발표하고 있으며, 메르세데스 벤츠는 올해 출시될 7세대 S클래스에 레벨3 자율주행기술을 탑재할 예정입니다. 

국내에는 7월부터 자율주행 레벨3 챠량의 판매가 허가되는데요. 정부는 ‘2030 미래 산업 발전전략’에서 2024년에 자율주행 기술 4단계 출시를 추진한다고 밝히며, 기술개발 추진 전략을 단계적 추진전략(레벨3->레벨4)에서 동시 추진전략(레벨3+레벨4)로 전환하여 2027년, 주요도로에 안전자율주행을 세계 최초로 상용화하겠다고 목표를 밝혔습니다
